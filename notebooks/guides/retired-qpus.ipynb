## Stillgelegte Prozessorfamilien
- **Canary**: Die Canary-Familie umfasste kleine Designs mit 5 bis 16 Qubits.
- **Falcon**: Die Falcon-Gerätefamilie bot eine wertvolle Plattform für mittelgroße Circuits und diente zudem als Grundlage, um Leistungs- und Skalierbarkeitsverbesserungen zu demonstrieren, bevor diese auf größere Geräte übertragen wurden.
- **Egret**: Egret brachte die Innovationen abstimmbarer Koppler auf eine 33-Qubit-Plattform und ermöglichte damit schnellere Zwei-Qubit-Gates mit höherer Genauigkeit.
- **Hummingbird**: Mit einem heavy-hexagonalen Qubit-Layout erlaubte die Hummingbird-Familie bis zu 65 Qubits.
- **Osprey**: Osprey war mit 433 Qubits fast viermal so groß wie Eagle. Die größeren Chip-Abmessungen erforderten weitere Verbesserungen bei der Geräteverpackung sowie maßgefertigte Flexkabel im Kryostaten, um die höheren Ein-/Ausgabeanforderungen innerhalb desselben Verkabelungsbereichs zu erfüllen.
- **Eagle**: Die Eagle-Prozessorfamilie integrierte im Vergleich zu früheren Generationen skalierbarere Verpackungstechnologien. Insbesondere wurden Signale durch mehrere Chipschichten geführt, was eine hohe I/O-Dichte ohne Leistungseinbußen ermöglichte.
